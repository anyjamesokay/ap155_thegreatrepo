```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, np.pi*2, 100)
f_1 = 3
f_2 = -5
phi = np.pi/2
psi = np.pi/4
y_1 = np.sin((f_1*x)+phi)
y_2 = np.sin((f_2*x)+psi)

fig, ax = plt.subplots(figsize=(4, 3))
ax.set_title('Sinusoidal Waves')



plt.plot(x, y_1*y_2)
plt.show()
```

```python
x = np.linspace(0, 1, 100)
y = np.sqrt(1-x**2)
z = np.random.multivariate_normal(mean=[0,0], cov=[[1,0.3],[0.3,1]],size=500)
n = []
fig, ax = plt.subplots(figsize=(4, 3), layout='constrained')

for i in z:
    k = i[0]
    l = i[1]
    if  k**2 + l**2 < 1 and k > 0 and l > 0:       
        n.append(i)
m = np.array(n)
for j in m:
    j = np.array(j)
ax.set_title('Quarter Circle')
ax.scatter(m[:,0], m[:,1])
plt.plot(x, y, color='red')
```

```python
b = np.random.normal(0.5, 0.15, 300)
c = []
for i in b:
    if 0 < i < 1:
        c.append(i)

c = np.array(c)
fig, ax = plt.subplots(figsize=(4, 3))
ax.set_title('Histogram')
ax.hist(c, 30)

```
