```python
import random

#A1. Create a function random_integer_matrix(nrows, ncols) that generates a matrix in the form of "list-of-lists" 
#    whose elements are random integers from 1 to 9. Each "sub-list" inside the main list is a row in the matrix.
def random_integer_matrix(nrows, ncols):
    i = 0
    j = 0
    k = 0
    rows = []
    matrix = []
    while j < nrows: 
        rows = []      #Empties the row list every iteration.
        i = 0          #Resets the inner loop index.
        while i < ncols:
            k = random.randint(1 , 9)
            rows.append(k)    # Appends random number into a row.
            i += 1
        
        matrix.append(rows)   #Fills up bigger list with these "rows" lists.
        j += 1   
    return matrix

#A2. Create a function scalar_product(list1, list2) where you multiply each element at the same index 
# in the two lists and sum/accumulate the results.
def scalar_product(list1, list2):
    sum = 0
    k = 0
    i = 0
    while i < len(list1):   #Knowing the lengths of list1 and list2 need to be equal, this is fine.
        k = list1[i] * list2[i] 
        sum += k
        i += 1
    return sum

#A3. Create a function transpose_matrix(orig_matrix) which returns the transpose of any matrix also in 
# "list-of-lists" form. Do not assume the size of the matrix; figure it out within the function.
def transpose_matrix(orig_matrix):
    new_rows = []
    new_matrix = []
    ncols = len(orig_matrix)                    # Getting the dimensions of the original matrix.
    nrows = len(orig_matrix[0])
    i = 0
    j = 0
    while j < nrows:           # Same iteration method as A1, but with the row and column dimensions swapped.
        new_rows = []
        i = 0
        while i < ncols:
            new_rows.append(orig_matrix[i][j])      # Gets the jth element from every row, 
            i += 1                                  # then appends it into a new column.
        new_matrix.append(new_rows)
        j += 1
    return new_matrix

#A4. Create a function matrix_multiply(left_matrix, right_matrix) which performs the matrix multiplication.
# Do not assume the matrices follow a specific shape.
def matrix_multiply(left_matrix, right_matrix):
    i = 0
    j = 0
    rows = []
    matrix = []
    left_matrix_ncols = len(left_matrix[0])
    left_matrix_nrows = len(left_matrix)
    right_matrix_ncols = len(right_matrix[0])
    right_matrix_nrows = len(right_matrix)
    if left_matrix_ncols == right_matrix_nrows: # Condition for matrix multiplication to work.
        right_transpose = transpose_matrix(right_matrix) # Transposes the right matrix so that scalar_product
        while j < left_matrix_nrows:                     # can smoothly be applied.
            rows = []
            i = 0
            while i < right_matrix_ncols:
                rows.append(scalar_product(left_matrix[j], right_transpose[i]))
                i += 1
            matrix.append(rows)
            j += 1
        return matrix
    else:
        return "Invalid matrix sizes!"

#Testing all 4 defined functions. Scalar_product is tested with a chosen length. test and test2 are the main subjects.
test = random_integer_matrix(3, 5)
test2 = random_integer_matrix(5, 2)
test3 = []
test4 = []
lengthh = 4
i = 0
while i < lengthh:
    test3.append(random.randint(1, 9))
    test4.append(random.randint(1, 9))
    i += 1


print(test)
print(test2)
print(test3)
print(test4)
print(scalar_product(test3, test4))
print(transpose_matrix(test))
print(matrix_multiply(test, test2))   #The most important test!
```
