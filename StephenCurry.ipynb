```python
import numpy as np
import matplotlib.pyplot as plt

#I HAVEN'T WATCHED KPOP DEMON HUNTERS I DON'T WANNA HEAR NO "GOLDEN"
#SO I'LL REFERENCE THE NBA TEAM INSTEAD
#P.S. I'M A LEBRON FAN.
def gsw(phi, x0, x1, kmax=200, tol=1.e-8):
    varphi = 0.5*(1 + np.sqrt(5))
    for k in range(1,kmax):
        x2 = x0 + (x1-x0)/(varphi+1)
        x3 = x0 + (x1-x0)*varphi/(varphi+1)
        if phi(x3) < phi(x2):
            x0 = x2
        else:
            x1 = x3
        xnew = (x0+x1)/2
        xdiff = abs(x1-x0)        
        if abs(xdiff) < tol:
            break
        else:
            xnew = None
    return xnew
def phi_1(x):
    return x**4 - 9*x**3 + 25*x**2 - 24*x + 4
def dphi_1(x):
    return 4*x**3 - 27*x**2 + 50*x - 24

def phi_2(x):
    return (-x**5) + (4*x**4) - (4*x**3) + ((x**2)*np.exp(x)) - \
            (2*x**2) - (4*x*np.exp(x)) + (8*x) + (4*np.exp(x)) - 8
def dphi_2(x):
    return (-5*x**4) + (16*x**3) -(12*x**2) - (2*x*np.exp(x))+(np.exp(x)*x**2)-(4*x)+8
    
global1 = gsw(phi_1, -15, 15)
global2 = gsw(phi_2, -15, 15)
x1 = np.arange(-6, 7, 0.1)
y0 = np.arange(-67, 30, 0.1)
y1, y2, dy1, dy2 = [], [], [], []
for t in x1:
    y1.append(phi_1(t))
    y2.append(phi_2(t))
    dy1.append(dphi_1(t))
    dy2.append(dphi_2(t))

xax = np.zeros(np.size(x1))

fig, axs = plt.subplots(1, 2, dpi=100)
axs[0].plot(x1, np.array(y1), label=r'$\phi_1(x)$')
axs[0].plot(x1, np.array(dy1), 'g-.', label=r'$\frac{d\phi_1}{dx}$')
axs[0].plot(x1, xax, 'k--')
axs[0].plot(global1*np.ones(np.size(y0)), y0, 'r--')
axs[0].set_ylim(-20, 9)
axs[0].fill_between(x1, -21*np.ones(np.size(x1)),10*np.ones(np.size(x1)), color='orange', alpha=0.4)
axs[0].scatter(global1, phi_1(global1), color='red', label='Global Minima')
axs[1].plot(x1, np.array(y2), label=r'$\phi_2(x)$')
axs[1].plot(x1, np.array(dy2), 'g-.', label=r'$\frac{d\phi_2}{dx}$')
axs[1].plot(x1, xax, 'k--')
axs[1].plot(global2*np.ones(np.size(y0)), y0, 'r--')
axs[1].set_ylim(-67, 16.7)
axs[1].fill_between(x1, -67*np.ones(np.size(x1)),20*np.ones(np.size(x1)), color='orange', alpha=0.4)
axs[1].scatter(global2, phi_2(global2), color='red', label='Global Minima')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[0].legend()
axs[1].legend()
fig.tight_layout()
plt.title('Global Minima of Functions using Golden-section Search', loc='right')
```
