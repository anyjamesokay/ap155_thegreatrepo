```python
import numpy as np
a = 1
b = 1e8
c = 1

xmin = (-b - np.sqrt(b**2 - 4*a*c))/2*a
xmax = (-b + np.sqrt(b**2 - 4*a*c))/2*a
print(xmin)
print(xmax)
print(xmin*xmax)
xmax2 = c/(a*xmin)
print(xmax2)
print(xmax2-xmax)
```
